```python
import ddh
from datetime import datetime
import numpy as np
import pandas as pd
from IPython.display import clear_output

ddh.load('datacatalog.worldbank.org')
```

```python
fields = ['nid','title','field_wbddh_data_class','field_wbddh_release_date','field_wbddh_modified_date','created','changed','field_ddh_harvest_src','field_wbddh_data_type']
df = pd.DataFrame(columns=fields).astype(int)

n = 0

try:
    for nid,rec in ddh.dataset.search(fields):
        row = []
        for field in fields:
            row.append(ddh.dataset.field(rec, field))
        
        
        df.loc[len(df)] = row
        n += 1
        if n % 100 == 0:
            clear_output(wait=True)
            print('Processed: {}'.format(n))
                
except ddh.APIError as err:
    print('API Error: {}'.format(err.url))

clear_output(wait=False)    

# there might be a more elegant way to do this
df['year_month'] = pd.DatetimeIndex(df['created']).year*100+pd.DatetimeIndex(df['created']).month
```

```python
df['year_month'] = pd.DatetimeIndex(df['created']).year*100+pd.DatetimeIndex(df['created']).month
df.pivot_table(values='nid',index='year_month',columns='field_wbddh_data_type',margins=True,aggfunc='count').replace(np.nan,'')
```
